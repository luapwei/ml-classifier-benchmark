## Übersicht der Machine-Learning-Algorithmen (Mini-Keras Strategien)


#### 1. Decision Tree Classifier
* **Prinzip:** Ein Basis-Klassifikator, der Daten durch eine Abfolge von binären Entscheidungsregeln (Baumstruktur) trennt.
* **Rolle:** Dient als Baseline-Modell für den initialen Vergleich des ML-Workflows.

#### 2. Random Forest Klassifikator (Ensemblemethode)
* **Prinzip:** Kombiniert mehrere Entscheidungsbäume zu einem starken Modell, um Overfitting zu vermeiden und die Robustheit zu erhöhen.
* **Besonderheit:** Nutzt das Prinzip des "Bagging", um durch Mitteilung vieler Bäume eine stabilere Vorhersage zu treffen.

#### 3. Fisher’sche lineare Diskriminanzanalyse (LDA)
* **Prinzip:** Ein klassisches Verfahren, das die Merkmale so transformiert, dass die Trennbarkeit zwischen den Klassen maximiert wird.
* **Fokus:** Optimiert das Verhältnis von Varianz zwischen den Klassen zur Varianz innerhalb der Klassen.

#### 4. SVM mit RBF Kernel (Radial Basis Function)
* **Prinzip:** Eine Support Vector Machine, die Daten mithilfe eines Gauß-Kernels in einen unendlich-dimensionalen Raum projiziert.
* **Stärke:** Ermöglicht die Modellierung hochgradig nicht-linearer Entscheidungsgrenzen.

#### 5. SVM mit linearem Kernel
* **Prinzip:** Sucht nach der optimalen Hyperebene (einer geraden Grenze), die den Spielraum (Margin) zwischen den Klassen maximiert.
* **Einsatz:** Effektiv, wenn die Daten bereits linear gut trennbar sind.

#### 6. SVM mit Polynomkern (3. Grad)
* **Prinzip:** Verwendet eine polynomielle Kernfunktion dritten Grades, um komplexere Zusammenhänge abzubilden.
* **Konfiguration:** Erfordert die explizite Festlegung des Parameters `degree=3`.

#### 7. Lineare SVM (LinearSVC)
* **Prinzip:** Eine spezialisierte Implementierung der SVM für lineare Probleme, die auf einer anderen mathematischen Optimierung (Liblinear) basiert.
* **Vorteil:** Skaliert oft besser bei größeren Datensätzen als die Standard-SVC mit linearem Kernel.

#### 8. XGBoost (Extreme Gradient Boosting)
* **Prinzip:** Eine State-of-the-art Ensemblemethode, bei der schwache Lernmodelle (Bäume) sequenziell aufgebaut werden, um Fehler der Vorgänger zu korrigieren.
* **Rolle:** Stellt die leistungsfähigste Strategie in der "Mini-Keras"-Bibliothek dar.

## Mapping-Tabelle: Algorithmus zu Strategy-Klasse

## Vorgegebener Code mit Annotationen

```python
# Codeteil 0
# a)
import numpy as np
# b)
import pandas as pd
# c)
import seaborn as sns
# d)
import matplotlib.pyplot as plt
# e)
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix


# Codeteil 1 a) (Init)
np.random.seed(123) #sorgt dafür, dass random funktionen immer das Gleiche liefern
# Codeteil 1 b)
iris = sns.load_dataset('iris') #lädt den Blumen-Datensatz
# Codeteil 1 c)
indices = np.random.permutation(len(iris)) #Radomized die Zeilennummern des Datensatzes


# Codeteil 2 a) (Datensatz aufteilen)
test = iris.iloc[indices[:15]] # integer-based location → slicing (Nimmt alles vom 1. integer bis zum 15.) = Test des Trainings
# Codeteil 2 b)
train = iris.iloc[indices[15:]] # Nimm alles vom 15. integer bis zum Ende = Trainingsdaten


# Codeteil 3 a) (Vorgaben)
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width'] # Messwerte
target = 'species' # Ausgabe (Soll vorhergesagt werden)
# Codeteil 3b) (Training)
clf = DecisionTreeClassifier(random_state=123) # clf=classifier, Objekt eines Entscheidungsbaums
clf.fit(train[features], train[target]) # Trainingsbefehl: KI soll anhand der features das target vorhersagen
# Codeteil 3c) (Prüfung)
y_pred = clf.predict(test[features]) # .predict = Vorhersage des target nur anhand der features (15 Test Daten die sie noch nicht kennt)


# Codeteil 4a) (Fehlerrate berechnen)
error_rate = np.mean(y_pred != test[target]) # Vergleicht vorhersage mit realität (Liste mit True/False) => Durch .mean Prozentanteil
print("Fehlklassifikationsrate:", error_rate)


# Codeteil 5a)
cm = confusion_matrix(test[target], y_pred) #Vergleicht Antworten mit den Vermutungen → Erstellt eine Matrix
labels = sorted(set(test[target])) #Namen der Blumenarten (Setosa, Versicolor, Virginica) alphabetisch sortiert
# In DataFrame umwandeln durch pandas
cm_df = pd.DataFrame(
 cm,
 index=labels, # Reference (wahre Klassen in Zeilen)
 columns=labels # Prediction (vorhergesagte Klassen in Spalten)
)
# Codeteil 5b) Tabelle wird farbiges Bild
print(cm_df)
plt.figure(figsize=(8, 6)) # Breite und Höhe in Zoll
sns.heatmap( #Hauptbefehl der Seaborn-Bibliothek
 cm_df, #Datenquelle (wurde oben mit pandas erstellt)
 annot=True, # Zahlen anzeigen
 fmt="d", # Zahlen in Dezimal
 cmap="Blues", # ähnlich zu white → skyblue (Farbpalette)
 cbar=True,
     linewidths=0.5,
 linecolor="white"
)
plt.title("Konfusionsmatrix: ")
plt.xlabel("Vorhergesagte Klasse")
plt.ylabel("Wahre Klasse")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
